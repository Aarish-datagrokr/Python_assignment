# 5. Write a separate function which take input data as values under the column name "phone" and processes it to return equivalent ASCII char of the value taken 2 at a time from left to right. Considering the value should be in between 65 to 99. If two-digit number is less than 65 replace it with O.

```python
def filter_phone(phone):
    phone = re.sub('[^0-9]','',phone)
    result = ''
    if len(phone) < 10:
        for i in range(0,len(phone),2):
            pair_of_two = int(phone[i]+phone[i+1])

            if pair_of_two <65:
                result+='O'
            else:
                result+=chr(pair_of_two)
    else:
        for i in range(0,len(phone)-1,2):
            pair_of_two = int(phone[i]+phone[i+1])

            if pair_of_two <65:
                result+='O'
            else:
                result+=chr(pair_of_two)

    return result 

    
def to_ascii():
    for i in range(len(df)):
        df.loc[i, 'phone'] = filter_phone(df.loc[i,'phone'])
    
    print(df['phone'])
    
to_ascii()    
```
